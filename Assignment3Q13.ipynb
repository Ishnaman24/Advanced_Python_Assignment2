```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from scipy.cluster.hierarchy import dendrogram, linkage
from scipy.spatial.distance import pdist

# Set seed for reproducibility
np.random.seed(42)

# Generate a random dataset
num_rows = 500
num_columns = 10

# Define column value ranges
column_ranges = [
    (-10, 10),
    (-10, 10),
    (-10, 10),
    (-10, 10),
    (10, 20),
    (10, 20),
    (10, 20),
    (10, 20),
    (-100, 100),
    (-100, 100)
]

# Generate random dataset
dataset = np.zeros((num_rows, num_columns))
for i, (min_val, max_val) in enumerate(column_ranges):
    dataset[:, i] = np.random.uniform(low=min_val, high=max_val, size=num_rows)

# Apply K-Means clustering and determine optimal number of clusters
inertia_values = []
range_clusters = range(1, 11)
for num_clusters in range_clusters:
    kmeans = KMeans(n_clusters=num_clusters)
    kmeans.fit(dataset)
    inertia_values.append(kmeans.inertia_)

# Plot the inertia values to find the optimal number of clusters for K-Means
plt.plot(range_clusters, inertia_values, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.title('K-Means Clustering: Elbow Method')
plt.show()

# Apply K-Means clustering with the optimal number of clusters
optimal_num_clusters_kmeans = 3  # Adjust this based on the elbow method plot
kmeans = KMeans(n_clusters=optimal_num_clusters_kmeans)
kmeans.fit(dataset)

# Apply Hierarchical clustering and determine optimal number of clusters
linkage_values = []
for num_clusters in range_clusters:
    linkage_matrix = linkage(dataset, method='ward')
    distances = pdist(dataset)
    linkage_values.append(linkage_matrix[-num_clusters, 2])

# Plot the linkage values to find the optimal number of clusters for Hierarchical
plt.plot(range_clusters, linkage_values, marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('Linkage Distance')
plt.title('Hierarchical Clustering: Linkage Distance')
plt.show()

# Apply Hierarchical clustering with the optimal number of clusters
optimal_num_clusters_hierarchical = 2  # Adjust this based on the linkage distance plot
hierarchical = AgglomerativeClustering(n_clusters=optimal_num_clusters_hierarchical)
hierarchical.fit(dataset)
distances = pairwise_distances_argmin_min(dataset, metric='euclidean')[1]
linkage_matrix = np.column_stack([distances, agg_clustering.labels_])
plt.figure(figsize=(10, 5))
plt.title('Hierarchical Clustering: Dendrogram')
dendrogram = plt.dendrogram(linkage_matrix)
plt.xlabel('Samples')
plt.ylabel('Distance')
plt.show()

```
