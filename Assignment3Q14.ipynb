```python
import numpy as np
import matplotlib.pyplot as plt

# Set seed for reproducibility
np.random.seed(42)

# Generate a random dataset
num_rows = 600
num_columns = 15
column_ranges = (-100, 100)
dataset = np.random.uniform(low=column_ranges[0], high=column_ranges[1], size=(num_rows, num_columns))

#  Plot scatter graph between Column 5 and Column 6
plt.scatter(dataset[:, 4], dataset[:, 5])
plt.xlabel('Column 5')
plt.ylabel('Column 6')
plt.title('Scatter Plot')
plt.show()

#  Plot histogram of each column in single graph
plt.figure(figsize=(10, 6))
for i in range(num_columns):
    plt.hist(dataset[:, i], bins=20, alpha=0.5, label=f'Column {i+1}')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Columns')
plt.legend()
plt.show()

#  Plot the Box plot of each column in single graph
plt.figure(figsize=(10, 6))
plt.boxplot(dataset)
plt.xlabel('Column')
plt.ylabel('Value')
plt.title('Box Plot of Columns')
plt.show()
```
