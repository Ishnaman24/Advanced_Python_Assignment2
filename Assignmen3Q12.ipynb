```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set seed for reproducibility
np.random.seed(42)

# Generate a random dataset
num_rows = 100
num_columns = 30
dataset = pd.DataFrame(np.random.uniform(low=1, high=200, size=(num_rows, num_columns)))

#  Replace values with NA in the defined range [10, 60]
dataset.loc[10:60, :] = np.nan

# Print the count of rows with missing values
missing_rows_count = dataset.isna().sum(axis=1).sum()
print("Number of rows with missing values:", missing_rows_count)

#  Replace NA values with column mean
dataset = dataset.fillna(dataset.mean())

#  Calculate Pearson correlation and plot heat map
cor_matrix = dataset.corr()
sns.heatmap(cor_matrix, annot=True, cmap="coolwarm")
plt.show()

# Select columns with correlation <= 0.7
cor_threshold = 0.7
columns_to_select = cor_matrix.columns[cor_matrix.abs().le(cor_threshold).all()]

#  Normalize values between 0 and 10
dataset_normalized = (dataset - dataset.min()) / (dataset.max() - dataset.min()) * 10

#  Replace values with 1 if <= 0.5, else with 0
dataset_binary = dataset_normalized.applymap(lambda x: 1 if x <= 0.5 else 0)
```
